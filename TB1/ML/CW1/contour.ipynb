```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

mu = np.array([-1.5,0.3])
K = np.eye(2)+np.matrix([[0,-1/2],[-0.5,0]])
N = 300
pdf = multivariate_normal(mu,K)
# generate points along axis
x = np.linspace(-3,3,N)
y = np.linspace(-3,3,N)
# get all combinations of the points
x1p,x2p = np.meshgrid(x,y)
pos = np.vstack((x1p.flatten(),x2p.flatten()))
pos = pos.T
# evaluate pdf at points
Z = pdf.pdf(pos)
Z = Z.reshape(N,N)
# plot contours
fig, ax = plt.subplots()
pdf_c = ax.contour(x1p,x2p,Z,3,colors='r')
```
