```python
import numpy as np
from scipy.stats import norm
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
import random
from matplotlib import cm
```

```python
def posterior(m0,S0,X,y,beta):
    S0_inv = np.linalg.inv(S0)
    sn = np.linalg.inv(S0_inv+beta*np.dot(np.transpose(X),X))
    aux = np.matrix([np.matmul(S0_inv,m0)])
    aux = np.transpose(aux)
    aux += beta*np.matmul(np.matrix.transpose(X),y)
    mn = np.matmul(sn,aux)
    return norm.pdf(mn,sn),mn,sn
```

```python
#parameters to generate data
w0 = 0.3
w1 = -1.5
```

```python
#generate some data
x_pts = np.linspace(-1,1,199)
random.shuffle(x_pts)
y_pts = np.ones(199)
for i in range(199):
    y_pts[i] = w1*x_pts[i]+w0+np.random.normal(0,0.03)
```

```python
#now lets define our prior
m0 = np.array([0,0])
S0 = np.eye(2)

#p(w) = normal(priormu,priorSigma)
prior_w = norm.pdf(m0,S0)
```

```python
#Create grid and multivariate normal
X, Y = np.meshgrid(x_pts,x_pts)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal(m0, S0)

#Make a 3D plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

```python
# create figure
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
beta = 1/0.03
#we already shuffled the x points so that
#we can use consecutive points now
#newx = x_pts[0]
#newy = y_pts[0]
#x = np.array([x_pts[0]])
#X = np.matrix([[1,x]])
#y = np.matrix([y_pts[0]])
#plt.plot(newx,newy,'rx')
#post = posterior(m0,S0,X,y,beta)
#slope = post[1][1]
#intercept = post[1][0]
colours = ['y','g','b']
fig, ax = plt.subplots(3, 2, sharex='col', sharey='row')
#axs = _axs.flatten()

# We are now going to add successive points and see how the posterior changes and to see what lines 
# we have achieved by 'learning'
for i in range(0,3):
    newx = x_pts[i]
    newy = y_pts[i]
    x = np.array(x_pts[0:i+1])
    X = np.array([np.ones(x.shape),x])
    X = np.matrix.transpose(X)
    y = np.matrix(y_pts[0:i+1])
    y = np.transpose(y)
    ax[i,0].plot(x_pts[0:i+1],y_pts[0:i+1],'rx')
    post = posterior(m0,S0,X,y,beta)
    slope = post[1][1]
    intercept = post[1][0]
    ordinates = slope*x_pts+intercept*np.ones(x_pts.shape[0])
    ordinates = np.transpose(ordinates)
    ax[i,0].plot(x_pts,ordinates,colours[i],label=str(i+1)+' data point(s)')
    ax[i,0].legend(loc='upper left')
    #plot posterior
    N = 600
    mean = np.squeeze(np.asarray(post[1]))
    pdf = multivariate_normal(mean,post[2])
    # generate points along axis
    x = np.linspace(-3,3,N)
    y = np.linspace(-3,3,N)
    # get all combinations of the points
    x1p,x2p = np.meshgrid(x,y)
    pos = np.vstack((x1p.flatten(),x2p.flatten()))
    pos = pos.T
    # evaluate pdf at points
    Z = pdf.pdf(pos)
    Z = Z.reshape(N,N)
    # plot contours
    #fig, ax = plt.subplots(nrows=3,ncols=1)
    pdf_c = ax[i,1].contour(x1p,x2p,Z,4,colors='r')
    ax[i,1].set_title("Posterior after " + str(i+1)+ " data point(s)")
```

```python
#plot prior
N = 300
pdf = multivariate_normal(m0,S0)
# generate points along axis
x = np.linspace(-3,3,N)
y = np.linspace(-3,3,N)
# get all combinations of the points
x1p,x2p = np.meshgrid(x,y)
pos = np.vstack((x1p.flatten(),x2p.flatten()))
pos = pos.T
# evaluate pdf at points
Z = pdf.pdf(pos)
Z = Z.reshape(N,N)
# plot contours
fig, ax = plt.subplots(nrows=1,ncols=2)
pdf_c = ax[0].contour(x1p,x2p,Z,3,colors='r')
ax[0].set_title("The chosen prior")

```
